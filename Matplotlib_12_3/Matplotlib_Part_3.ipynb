# Part 3: Pie Charts

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Vẽ biểu đồ tròn

```python
slices = [120, 80, 30, 20]
labels = ['One hundred twenty', 'Eighty', 'Thirty', 'Twenty']
colors = ['grey', 'orange', 'pink', 'red']
plt.style.use('fivethirtyeight')
# wedgeprops={'edgecolor': 'black'}: đường bao quanh và phân cách các phần của biểu đồ
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, colors=colors)
plt.title('My Awesome Pie Chart')
plt.tight_layout()
```

##### Cho dữ liệu 1

```python
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
slices = [59219, 55466, 47544, 36443, 35917]
```

```python
plt.style.use('fivethirtyeight')
# explode=explode: chỉ định phần nào nhô ra trong biểu đồ
explode = [0.5, 0, 0, 0, 0]
# wedgeprops={'edgecolor': 'black'}: màu đường bao quanh và phân cách các phần của biểu đồ
# shadow=True: đổ bóng cho biểu đồ
# startangle=-10: xoay biểu đồ theo góc -10 độ
# autopct='%1.2f%%': hiển thị phần trăm của các phần trong biểu đồ, làm tròn đến chữ số thập phân thứ 2
plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'yellow'}, explode=explode, 
        shadow=True, startangle=-10, autopct='%1.2f%%')
plt.title('My Awesome Pie Chart')
plt.tight_layout()
```
