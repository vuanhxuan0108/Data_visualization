```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r"./data/data_1.csv")
```

```python
data.head()
```

```python
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color = '#444444', linestyle = "--", label='All Devs', alpha = 0.85)
plt.plot(ages, py_salaries, label = 'Python', color = '#999999')
# fill_between: hàm tô màu khu vực
# alpha: độ nhạt của màu
overall_median = dev_salaries # đường cắt ngang giới hạn biểu đồ phân chia 2 miền
# where(điều kiện): đường cắt dọc giới hạn biểu đồ phân chia 2 miền
# interpolate=True: khi sử dụng where và hai đường cong giao nhau. Nó tính toán điểm giao nhau thực tế và mở rộng khu vực đã tô màu đến điểm này 
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries > overall_median), alpha = 0.15, interpolate=True, color = 'black')
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries <= overall_median), color = 'red', alpha = 0.25, interpolate=True)
plt.title("Median Salary (USD) by Age")
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.legend()
plt.tight_layout()
```
